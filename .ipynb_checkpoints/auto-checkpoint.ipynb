```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("auto.csv")
```

```python
df.columns = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.shape
```

```python
df.isin(["?"]).sum()
```

```python
df.dtypes
```

```python
df.head(5)
```

```python
df.replace("?", np.nan , inplace= True)
```

```python
df["price"].unique()
```

```python
df["price"] = pd.to_numeric(df["price"].astype(str),errors="coerce")
```

```python
df["bore"] =  pd.to_numeric(df["bore"].astype(str),errors="coerce")  
df["stroke"] =  pd.to_numeric(df["stroke"].astype(str),errors="coerce")  
df["peak-rpm"] =  pd.to_numeric(df["peak-rpm"].astype(str),errors="coerce")  
df["horsepower"] =  pd.to_numeric(df["horsepower"].astype(str),errors="coerce")  
```

```python
df["peak-rpm"] = df["peak-rpm"].astype(float)
df["horsepower"] = df["horsepower"].astype(float)
```

```python
df.dropna(subset=["price"],inplace=True)
df.shape
```

```python
df.isna().sum()
```

```python
df["normalized-losses"].replace(np.nan , df["normalized-losses"].mean() , inplace=True )
```

```python
df.dtypes
```

```python
cols=["normalized-losses","bore","stroke","horsepower","peak-rpm","price"]
```

```python
df[cols] = pd.to_numeric(cols , errors="coerce")
```

```python
df.head() 
```

```python
df["bore"].replace(np.nan , df["bore"].mean() , inplace=True )
df["stroke"].replace(np.nan , df["stroke"].mean() , inplace=True )
df["horsepower"].replace(np.nan , df["horsepower"].mean() , inplace=True )
df["peak-rpm"].replace(np.nan , df["peak-rpm"].mean() , inplace=True )
df["price"].replace(np.nan , df["price"].mean() , inplace=True )
```
